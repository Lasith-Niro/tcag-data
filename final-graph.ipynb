```python
# plt.subplot()
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from pylab import cm
import pandas as pd
#http://olsgaard.dk/monochrome-black-white-plots-in-matplotlib.html#Line-and-marker-styles
from pylab import rcParams
rcParams['figure.figsize'] = 15, 10
df = pd.read_csv("uids.csv")
plt.style.use("seaborn-poster")
plt.style.context('bmh')
plt.get_cmap('jet') 
plt.xlabel("Year")
plt.ylabel("Distinct users")
plt.title("OSM Editor usage by number of users")

plt.plot(df["Year"], df['iD'],'--d')
plt.plot(df["Year"], df['JOSM'], '-D')
plt.plot(df["Year"], df['Potlatch2'], '-s')
plt.plot(df["Year"], df['MAPS.ME'], ':H')
plt.plot(df["Year"], df['StreetComplete'],'-X')



plt.legend(["iD","JOSM", "Potlatch 2", "MAPS.ME", "StreetComplete"], bbox_to_anchor=(1.05, 1),loc='upper left', 
           borderaxespad=0.)
plt.show()
```
