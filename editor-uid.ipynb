```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from pylab import cm
import pandas as pd
#http://olsgaard.dk/monochrome-black-white-plots-in-matplotlib.html#Line-and-marker-styles
from pylab import rcParams
rcParams['figure.figsize'] = 15, 10
```

```python
df = pd.read_csv("uids.csv")
```

```python
df
```

```python
from cycler import cycler
monochrome = (cycler('color', ['k']) * cycler('linestyle', ['-', '--', ':', '=.']) * cycler('marker', ['^',',', '.']))
```

```python
plt.style.use("seaborn-poster")
fig, ax = plt.subplots()
ax.set_prop_cycle(monochrome)
ax.set_xlabel("Year")
ax.set_ylabel("Distinct users")
ax.set_title("OSM Editor usage by number of users")


ax.plot(df["Year"], df['iD'])
ax.plot(df["Year"], df['JOSM'])
ax.plot(df["Year"], df['Potlatch 2'])
ax.plot(df["Year"], df['MAPS.ME'])
ax.plot(df["Year"], df['StreetComplete'])

# ax.set_xticks(ax.get_xticks()[::2])




ax.legend(["iD","JOSM", "Potlatch 2", "MAPS.ME","StreetComplete"], bbox_to_anchor=(1.05, 1),loc='upper left', borderaxespad=0.)
plt.show()
```

```python
# plt.subplot()

plt.style.use("seaborn-poster")
# plt.get_cmap('jet') 
plt.xlabel("Year")
plt.ylabel("Distinct users")
plt.title("OSM Editor usage by number of users")

plt.plot(df["Year"], df['iD'], '-P')
plt.plot(df["Year"], df['JOSM'], 'k--D')
plt.plot(df["Year"], df['Potlatch 2'], 'k-s')
plt.plot(df["Year"], df['MAPS.ME'], 'k:')
plt.plot(df["Year"], df['StreetComplete'],'--X')


plt.legend(["iD","JOSM", "Potlatch 2", "MAPS.ME","StreetComplete"], bbox_to_anchor=(1.05, 1),loc='upper left', borderaxespad=0.)
plt.show()
```

```python
plt.style.use("seaborn-poster")
plt.style.context('bmh')
plt.get_cmap('jet') 
plt.xlabel("Year")
plt.ylabel("Distinct users")
plt.title("OSM Editor usage by number of users")

plt.plot(df["Year"], df['iD'],'--d')
plt.plot(df["Year"], df['JOSM'], '-D')
plt.plot(df["Year"], df['Potlatch 2'], '-s')
plt.plot(df["Year"], df['MAPS.ME'], ':')


plt.legend(["iD","JOSM", "Potlatch 2", "MAPS.ME"], bbox_to_anchor=(1.05, 1),loc='upper left', borderaxespad=0.)
plt.show()
```
